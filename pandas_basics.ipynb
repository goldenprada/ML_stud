```python
d = {"No": False, "Yes": True}
df["International plan"] = df["International plan"].map(d)
df.head() #replacing values in a column via map method
```

```python
df = df.replace({"Voice mail plan": d}) #same thing with reduce
df.head()
```

```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"] #grouping

df.groupby(["Churn"])[columns_to_show].describe(percentiles=[])
```

```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]

df.groupby(["Churn"])[columns_to_show].agg([np.mean, np.std, np.min, np.max])
```

```python
pd.crosstab(df["Churn"], df["International plan"]) #summary tables
```

```python
pd.crosstab(df["Churn"], df["Voice mail plan"], normalize=True)
```

```python
df.pivot_table(
    ["Total day calls", "Total eve calls", "Total night calls"],
    ["Area code"],
    aggfunc="mean",
) #pivot table
```

```python
total_calls = (
    df["Total day calls"]
    + df["Total eve calls"]
    + df["Total night calls"]
    + df["Total intl calls"]
)
df.insert(loc=len(df.columns), column="Total calls", value=total_calls) #adding a column
```

```python
df.head()
```

```python
df["Total charge"] = (
    df["Total day charge"]
    + df["Total eve charge"]
    + df["Total night charge"]
    + df["Total intl charge"]
)
df.head() #another way to add a column
```

```python
df.drop(["Total charge", "Total calls"], axis=1, inplace=True)
```

```python
df.drop([1, 2]).head()
```

```python
pd.crosstab(df["Churn"], df["International plan"], margins=True)
```
